```python
import pymatgen as pymat
from pymatgen.core.periodic_table import Element
import mendeleev as mendel
import pandas as pd
import matplotlib.pyplot as plt

fcc_elements = ["Ag", "Al", "Au", "Cu", "Ir", "Ni", "Pb", "Pd", "Pt", "Rh", "Sr", "Th", "Yb"]
bcc_elements = ["Ba", "Cr", "Cs", "Eu", "Fe", "K", "Li", "Mn", "Mo", "Na", "Nb", "Rb", "Ta", "V", "W"]
hcp_elements = ["Be", "Ca", "Cd", "Co", "Dy", "Er", "Gd", "Hf", "Ho", "Lu", "Mg", "Os", "Re", "Ru", 
                "Sc", "Tb", "Tc", "Ti", "Tl", "Tm", "Y", "Zn", "Zr"]
other_elements = ["H", "He", "B", "N", "O", "F", "Ne", "Si", "Cl", "Ar"]

elements = fcc_elements + bcc_elements + hcp_elements + other_elements

querable_pymatgen = ["atomic_mass", "atomic_radius", "electrical_resistivity", "molar_volume", "thermal_conductivity", 
                     "bulk_modulus", "youngs_modulus", "brinell_hardness", "average_ionic_radius", "melting_point", 
                     "rigidity_modulus", "density_of_solid", "coefficient_of_linear_thermal_expansion"]
                    
querable_mendeleev = ["atomic_number", "atomic_volume", "boiling_point", "electron_affinity", "en_pauling", "evaporation_heat", 
                      "fusion_heat", "heat_of_formation", "lattice_constant", "melting_point", "specific_heat", "density", 
                      "molar_heat_capacity", "atomic_weight_uncertainty", "dipole_polarizability"]

```

```python
all_values = []

for item in elements:
    element_values = []

    element_object = Element(item)
    for i in querable_pymatgen:
        element_values.append(getattr(element_object, i))

    element_object = mendel.element(item)
    for i in querable_mendeleev:
        element_values.append(getattr(element_object, i))

    all_values.append(element_values)
                      
```

```python
df = pd.DataFrame(all_values, columns = querable_pymatgen + querable_mendeleev)
display(df)

df.index = elements
display (df)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(axis = 1, how = 'any')
print (df)
```

```python
display (df)
```

```python
y = df.iloc[:,4]
print (y)
```

```python
import numpy as np
X = df.iloc[:,np.r_[0:4, 5:10]]
display(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 66)
```

```python
display(X_test)
```

```python
display(y_test)
```

```python
display(X_train)
```

```python
display(y_train)
```

```python
from sklearn.linear_model import Lasso
import numpy as np 

lasso = Lasso().fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso.score(X_test, y_test)))
print("Number of features used: {}".format(np.sum(lasso.coef_ !=0)))
```

```python
lasso01 = Lasso(alpha = 0.1, max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso01.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso01.score(X_test, y_test)))
print("Number of features used: {}".format(np.sum(lasso01.coef_ !=0)))
```
